판다스에는 DATA Frame과  Series라는 두가지가 있다

```python
pd.DataFrame({'Yes': [50, 21], 'No': [131, 2]})
```

DATA Frame은 테이블이다. 위 코드를 행렬로 생각한다면 1열은 Yes, 2열은 No가 되며

1행은(인덱스값은 0) 50, 131 이 각각 1행 1열 2열에 들어가고

2행은 21, 2가 각각 2행 1열 2열에 들어간다

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 'Sue': ['Pretty good.', 'Bland.']})
```

DATAFrame은 정수로 제한이 안되고 문자열도 가능하다

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 
              'Sue': ['Pretty good.', 'Bland.']},
             index=['Product A', 'Product B'])
```

이렇게 하게 된다면 위에서 봤던 1행의 인덱스 값은 0이 아닌 Product A로 바뀌게 된다

```python
pd.Series([1, 2, 3, 4, 5])
```
